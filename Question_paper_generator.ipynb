```python
# ==========================================================
# GOOGLE COLAB NOTEBOOK
# AI EXAMINATION PAPER GENERATION SYSTEM
# USING GROQ + LLAMA 3
# ==========================================================

# ==========================================================
# STEP 1: INSTALL REQUIRED LIBRARIES
# ==========================================================

!pip install -q groq fpdf

# ==========================================================
# STEP 2: IMPORT LIBRARIES
# ==========================================================

import random
from groq import Groq
from fpdf import FPDF
from google.colab import userdata

# ==========================================================
# STEP 3: LOAD GROQ API KEY
# ==========================================================

# In Google Colab:
# Left Sidebar → Secrets → Add New Secret
# Name: GROQ_API_KEY
# Value: Your Groq API Key

GROQ_API_KEY = userdata.get('GROQ_API_KEY')

# Initialize Groq Client
client = Groq(api_key=GROQ_API_KEY)

# ==========================================================
# STEP 4: SAMPLE SYLLABUS
# ==========================================================

syllabus = {
    "Unit 1": [
        "Introduction to Operating Systems",
        "Types of Operating Systems",
        "Functions of Operating Systems"
    ],

    "Unit 2": [
        "CPU Scheduling Algorithms",
        "Deadlock",
        "Process Synchronization"
    ],

    "Unit 3": [
        "Memory Management",
        "Paging and Segmentation",
        "Virtual Memory"
    ]
}

# ==========================================================
# STEP 5: BLOOM TAXONOMY LEVELS
# ==========================================================

blooms_levels = [
    "Remember",
    "Understand",
    "Apply",
    "Analyze",
    "Evaluate",
    "Create"
]

# ==========================================================
# STEP 6: DIFFICULTY LEVELS
# ==========================================================

difficulty_levels = [
    "Easy",
    "Medium",
    "Hard"
]

# ==========================================================
# STEP 7: QUESTION GENERATION FUNCTION
# ==========================================================

def generate_questions(unit, topic, difficulty, bloom_level):

    prompt = f"""
    You are an intelligent AI examination paper generator.

    Generate:
    1. Two MCQs
    2. Two descriptive questions
    3. Correct answer keys

    Constraints:
    - Unit: {unit}
    - Topic: {topic}
    - Difficulty: {difficulty}
    - Bloom Taxonomy Level: {bloom_level}

    Format Properly.
    """

    try:

        response = client.chat.completions.create(
            model="llama-3.1-8b-instant",
            messages=[
                {
                    "role": "user",
                    "content": prompt
                }
            ],
            temperature=0.7,
            max_tokens=1000
        )

        return response.choices[0].message.content

    except Exception as e:
        return f"Error: {str(e)}"

# ==========================================================
# STEP 8: GENERATE COMPLETE QUESTION PAPER
# ==========================================================

def generate_question_paper():

    complete_paper = ""

    for unit, topics in syllabus.items():

        complete_paper += "\n"
        complete_paper += "=" * 60
        complete_paper += f"\n{unit}\n"
        complete_paper += "=" * 60 + "\n"

        for topic in topics:

            difficulty = random.choice(difficulty_levels)
            bloom_level = random.choice(blooms_levels)

            print(f"Generating Questions for: {topic}")

            generated_questions = generate_questions(
                unit,
                topic,
                difficulty,
                bloom_level
            )

            complete_paper += f"\nTOPIC: {topic}\n"
            complete_paper += f"Difficulty: {difficulty}\n"
            complete_paper += f"Bloom Level: {bloom_level}\n\n"

            complete_paper += generated_questions
            complete_paper += "\n\n"

    return complete_paper

# ==========================================================
# STEP 9: SAVE QUESTION PAPER AS PDF
# ==========================================================

def save_pdf(content, filename="AI_Question_Paper.pdf"):

    pdf = FPDF()
    pdf.set_auto_page_break(auto=True, margin=15)

    pdf.add_page()

    pdf.set_font("Arial", size=12)

    lines = content.split("\n")

    for line in lines:
        pdf.multi_cell(0, 8, line)

    pdf.output(filename)

    print(f"\nPDF Saved Successfully: {filename}")

# ==========================================================
# STEP 10: MAIN EXECUTION
# ==========================================================

print("=" * 60)
print("AI EXAMINATION PAPER GENERATION SYSTEM")
print("=" * 60)

question_paper = generate_question_paper()

print("\n\nGENERATED QUESTION PAPER\n")
print(question_paper)

# Save PDF
save_pdf(question_paper)
```
