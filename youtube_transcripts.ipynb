```python
import gradio as gr
from groq import Groq
import os

client = Groq(
    api_key="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"
)
def generate_learning_material(transcript):

    prompt = f"""
    Analyze the following educational content.
    Generate:
    1. Summary
    2. Key Concepts
    3. Flashcards
    4. Quiz Questions with Answers
    5. Learning Outcomes
    6. Beginner Friendly Explanation
    Content:
    {transcript}
    """

    response = client.chat.completions.create(

        model="llama-3.3-70b-versatile",

        messages=[
            {
                "role": "user",
                "content": prompt
            }
        ],

        temperature=0.9,
        max_tokens=2000
    )

    return response.choices[0].message.content

def analyze_transcript(transcript):

    try:

        learning_material = generate_learning_material(transcript)

        return learning_material

    except Exception as e:

        return str(e)

iface = gr.Interface(

    fn=analyze_transcript,

    inputs=gr.Textbox(
        lines=15,
        label="Paste YouTube Transcript or Educational Notes",
        placeholder="Paste transcript text here..."
    ),

    outputs=gr.Markdown(
        label="AI Generated Learning Materials"
    ),

    title="AI Learning Assistant",

    description="""
    Paste educational transcripts or notes.
    AI generates:
    • Summary
    • Key Concepts
    • Flashcards
    • Quiz Questions
    • Learning Outcomes
    Powered by:
    Groq + Hugging Face
    """
)

iface.launch()
```
