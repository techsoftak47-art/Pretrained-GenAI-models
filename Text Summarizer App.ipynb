```python
client = Groq(api_key=GROQ_API_KEY)
```

```python
def summarize_text(text, summary_type):

    try:

        if text is None or text.strip() == "":
            return "Please enter some text."

        # Prompt selection
        if summary_type == "bullet points":

            prompt = (
                "Convert the following text into clear bullet points.\n\n"
                f"Text:\n{text}"
            )

        else:

            prompt = (
                f"Summarize the following text in a {summary_type} way.\n\n"
                f"Text:\n{text}"
            )

        # API call
        response = client.chat.completions.create(

            model="llama-3.1-8b-instant",

            messages=[
                {
                    "role": "user",
                    "content": prompt
                }
            ],

            temperature=0.2,
            max_tokens=180
        )

        return response.choices[0].message.content

    except Exception as e:

        return "Error: " + str(e)
```

```python
demo = gr.Interface(

    fn=summarize_text,

    inputs=[

        gr.Textbox(
            lines=10,
            label="Enter Text",
            placeholder="Paste text here..."
        ),

        gr.Radio(
            ["short", "medium", "detailed", "bullet points"],
            value="medium",
            label="Summary Type"
        )
    ],

    outputs=gr.Textbox(
        lines=8,
        label="Generated Summary"
    ),

    title="AI Text Summarizer App",

    description="Paste text and generate AI summaries using Groq."
)
```

```python
demo.launch(share=True, debug=True)
```
